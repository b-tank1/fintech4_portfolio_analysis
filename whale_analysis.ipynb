# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp_tsx_history.csv`: Contains historical closing prices of the S&P TSX 60 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data.

```python
# Reading whale returns
whale_path = Path("Resources/whale_returns.csv")
whale_returns = pd.read_csv(whale_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
whale_returns.head()
```

```python
# Count nulls
whale_returns.isnull().sum()
```

```python
# Drop nulls
whale_returns = whale_returns.dropna()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data.

```python
# Reading algorithmic returns
algo_path = Path("Resources/algo_returns.csv")
algo_returns = pd.read_csv(algo_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
algo_returns.head()
```

```python
# Count nulls
algo_returns.isnull().sum()
```

```python
# Drop nulls
algo_returns = algo_returns.dropna()
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
tsx_path = Path("Resources/sp_tsx_history.csv")
tsx_data = pd.read_csv(tsx_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
tsx_data.head()
```

```python
# Check Data Types
tsx_data.dtypes
```

```python
#remove $ and , in price and convert to float
tsx_data['Close'] = tsx_data['Close'].str.translate({ord(i): None for i in '$,'})
tsx_data['Close'] = tsx_data['Close'].astype('float')
```

```python
# First Sort Prices by Date and then Calculate Daily Returns and Drop nulls
tsx_data = tsx_data.sort_index(ascending=True)
tsx_returns = tsx_data.pct_change().dropna().copy()
```

```python
# Rename `Close` Column to be specific to this portfolio.
tsx_returns.rename(columns = {'Close':'S&P TSX 60'}, inplace=True)
tsx_returns.head()
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
daily_returns = pd.concat(
    [whale_returns, algo_returns, tsx_returns], axis="columns", join="inner"
)
daily_returns.head()
```

## Risk Analysis

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios.
4. Determine which portfolios are riskier than the S&P TSX 60.
5. Calculate the Annualized Standard Deviation.

### Create a box plot for each portfolio


```python
# Box plot to visually show risk
daily_returns.plot.box(figsize=(25, 10), title="Daily portfolio returns")
```

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios
daily_returns_std = daily_returns.std()
```

### Determine which portfolios are riskier than the S&P TSX 60

```python
# Calculate  the daily standard deviation of S&P TSX 60
tsx_std = daily_returns_std['S&P TSX 60']

# Determine which portfolios are riskier than the S&P TSX 60
daily_returns_std.where(daily_returns_std > tsx_std).dropna()
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
daily_returns_std_annualized = daily_returns_std * np.sqrt(252)
daily_returns_std_annualized
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.
3. Choose one portfolio, then calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
rolling_std = daily_returns.rolling(window=21).std().dropna()

# Plot the rolling standard deviation
rolling_std.plot(figsize=(25, 10), title="21-day rolling std")
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation = daily_returns.corr()

# Display de correlation matrix
correlation
```

```python
#show correlation heatmap
sns.heatmap(correlation, vmin=-1, vmax=1)
```

Algo2 appears to be most correlated to the S&P TSX 60.

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of Algo1, 60-day rolling
rolling_algo1_cov = daily_returns["Algo 1"].rolling(window=60).cov(daily_returns["S&P TSX 60"])

# Calculate 60-day rolling vavariance of S&P TSX
rolling_variance = daily_returns["S&P TSX 60"].rolling(window=60).var()

# Computing beta
rolling_beta = rolling_algo1_cov / rolling_variance

# Plot beta trend
rolling_beta.plot(figsize=(25, 10), title="Rolling 60-Day Beta of Algo1")
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half life for each portfolio, using standard deviation (`std`) as the metric of interest.

```python
# Use `ewm` to calculate the rolling window
ewm_std = daily_returns.ewm(halflife=21).std()

ewm_std.plot(figsize=(25, 10), title="21-day exponentially weighted std")
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = (daily_returns.mean() * 252) / daily_returns_std_annualized
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot.bar(title="Sharpe Ratios")
```

### Determine whether the algorithmic strategies outperform both the market (S&P TSX 60) and the whales portfolios.

Algo 1 has a better Sharpe Ratio than all of the whale portfolios and the S&P TSX 60, so it yields a better return per unit of risk. Hence, Algo 1 outperforms the market and the whale portfolios.

Algo 2, on the other hand, does outperform the market but underperforms compared to Warren Buffet's fund, Berkshire Hathaway.

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Reading data from 1st stock, Shopify (2018-Jan to 2019-Dec)
shop_path = Path("Resources/shop_historical.csv")
shop_data = pd.read_csv(shop_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
shop_data.head()
```

```python
# Reading data from 2nd stock, Apple (2018-Jan to 2019-Dec)
aapl_path = Path("Resources/aapl_historical.csv")
aapl_data = pd.read_csv(aapl_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
aapl_data.head()
```

```python
# Reading data from 3rd stock, Nvidia (2018-Jan to 2019-Dec)
nvda_path = Path("Resources/nvda_historical.csv")
nvda_data = pd.read_csv(nvda_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
nvda_data.head()
```

```python
# Reading data from 4th stock, Costco (2018-Jan to 2019-Dec)
cost_path = Path("Resources/cost_historical.csv")
cost_data = pd.read_csv(cost_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
cost_data.head()
```

```python
# Reading data from 5th stock, Netflix (2018-Jan to 2019-Dec)
nflx_path = Path("Resources/nflx_historical.csv")
nflx_data = pd.read_csv(nflx_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
nflx_data.head()
```

```python
# Re-name each stock's close price with symbol
shop_data.rename(columns = {'Close':'SHOP'}, inplace=True)
aapl_data.rename(columns = {'Close':'AAPL'}, inplace=True)
nvda_data.rename(columns = {'Close':'NVDA'}, inplace=True)
cost_data.rename(columns = {'Close':'COST'}, inplace=True)
nflx_data.rename(columns = {'Close':'NFLX'}, inplace=True)

stock_prices = pd.concat(
    [shop_data, aapl_data, nvda_data, cost_data, nflx_data], axis="columns", join="inner"
)
stock_prices.head()
```

```python
# Calculate daily returns, first sorting by dates ascending, also dropping NAs
stock_prices = stock_prices.sort_index(ascending=True)
stock_returns = stock_prices.pct_change().dropna().copy()
stock_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/5, 1/5, 1/5, 1/5, 1/5]

# Calculate portfolio return, also converting to dataframe with name for Custom portfolio
portfolio_returns = stock_returns.dot(weights).to_frame('Custom Portfolio')

# Display sample data
portfolio_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
all_returns = pd.concat(
    [daily_returns, portfolio_returns], axis="columns", join="inner"
)
all_returns.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
all_returns.dropna(inplace=True)
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`
all_returns_std_annualized = all_returns.std() * np.sqrt(252)
all_returns_std_annualized
```

### Calculate and plot rolling `std` with 21-day window

```python
# Calculate rolling standard deviation
rolling_std_all = all_returns.rolling(window=21).std().dropna()

# Plot the rolling standard deviation
rolling_std_all.plot(figsize=(25, 10), title="21-day rolling std with custom portfolio")
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation_all = all_returns.corr()

# Display the correlation matrix
correlation_all
```

```python
#show correlation heatmap
sns.heatmap(correlation_all, vmin=-1, vmax=1)
```

### Calculate and Plot the 60-day Rolling Beta for Your Portfolio compared to the S&P 60 TSX

```python
# Calculate covariance of custom portfolio, 60-day rolling
rolling_custom_cov = all_returns["Custom Portfolio"].rolling(window=60).cov(all_returns["S&P TSX 60"])

# Calculate 60-day rolling vavariance of S&P TSX
rolling_variance = all_returns["S&P TSX 60"].rolling(window=60).var()

# Computing beta
custom_beta = rolling_custom_cov / rolling_variance

# Plot beta trend
custom_beta.plot(figsize=(25, 10), title="Rolling 60-Day Beta of Custom Portfolio")
```

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Calculate Annualized Sharpe Ratios
sharpe_ratios_all = (all_returns.mean() * 252) / all_returns_std_annualized
sharpe_ratios_all
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios_all.plot.bar(title="Sharpe Ratios with Custom Portfolio")
```

My custom portfolio consisting of SHOP, AAPL, NVDA, COST and NFLX does outperform the market.<br />
Because these are all notable large cap stocks, the custom portfolio is much more correlated to the market compared to Algo 1, given by the higher Beta. The custom portfolio also generally exudes higher volatility than all other portfolios, shown in the std plot, because it is less diversified. <br />
However, the five selected stocks are all some of the best performing large caps. Hence, the overall Sharpe ratio still outperforms all other portfolios except Algo 1.
